```python
import cv2
import numpy as np
```

```python
path = "big_arrows/created/cropped_object.jpg"

def get_corners(np_img):
    corners = cv2.goodFeaturesToTrack(np_img,
                                      maxCorners=7,
                                      qualityLevel=0.01,
                                      minDistance=5)

    corners = np.int32(corners)
    return corners

def create_img_np_array(path):
    img = cv2.imread(path)       # Reads as BGR by default
    img_array = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)
    return gray
```

```python
np_img = create_img_np_array(path)
approx = get_corners(np_img)
print(len(approx))
```
