```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = "big_arrows/created/cropped_object.jpg"

def get_corners(np_img):
    corners = cv2.goodFeaturesToTrack(np_img,
                                      maxCorners=7,
                                      qualityLevel=0.1,
                                      minDistance=2)

    corners = np.uint64(corners)
    return corners

def create_img_np_array(p):
    img = cv2.imread(p)       # Reads as BGR by default
    img_array = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # gray = cv2.cvtColor(img_array, cv2.COLOR_BGR2GRAY)
    return img_array

def render_dots(img, items):
    for i in items:
        img[i[0][1], i[0][0]] = 255
```

```python
np_img = create_img_np_array(path)
approx = get_corners(np_img)
val = approx.tolist()
render_dots(np_img, val)
cv2.imshow('asdf', np_img)
cv2.waitKey(0)
plt.imshow(np_img, interpolation='nearest', cmap='gray')
plt.show()
```
